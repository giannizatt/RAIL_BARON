# Play With Python

First of all, load the data!

```python
import csv
with open('../datasets/speedtest.csv', 'r') as csvreader:
    reader = csv.DictReader(csvreader)
    data = [obj for obj in reader]
print len(data)
print data[0]
```

## EX 1

How many times Turin appears? (on key: **CLIENT_CITY**)

```python
from collections import Counter

c = Counter([line['CLIENT_CITY'].lower() for line in data])
print c['turin']
```

```python
for k, v in c.items():
    print (k, v) 
```

## EX 4

Find entry with MIN Value of **DOWNLOAD_KBPS** and **UPLOAD_KBPS**

```python
## EX3 and EX4 are merged
dwnld = [int(line['DOWNLOAD_KBPS']) for line in data]
upld = [int(line['UPLOAD_KBPS']) for line in data]
print "Download kb/s: from %s to %s kb/s" % (min(dwnld), max(dwnld))
print "Upload kb/s: from %s to %s kb/s" % (min(upld), max(upld))
```

## EX 5 

Find mean velocity for **DOWNLOAD_KBPS** and **UPLOAD_KBPS** for Turin city

```python
dwnld_turin = [float(line['DOWNLOAD_KBPS']) for line in (filter(lambda x: x['CLIENT_CITY'].lower() == 'turin', data))]
upld_turin = [float(line['UPLOAD_KBPS']) for line in (filter(lambda x: x['CLIENT_CITY'].lower() == 'turin', data))]
print "Mean Download speed for Turin clients: %g kb/s" % (sum(dwnld_turin)/len(dwnld_turin))
print "Mean Upload speed for Turin clients: %g kb/s" % (sum(upld_turin)/len(upld_turin))
```

## EX 6

For each city, Find mean velocity for **DOWNLOAD_KBPS** and **UPLOAD_KBPS**

```python
from collections import defaultdict

velocities_up_dict = defaultdict(int)
velocities_down_dict = defaultdict(int)
data_len_by_city_dict = defaultdict(int)

for line in data:
    velocities_up_dict[line['CLIENT_CITY'].lower()] += float(line['UPLOAD_KBPS'])
    velocities_down_dict[line['CLIENT_CITY'].lower()] += float(line['DOWNLOAD_KBPS'])
    data_len_by_city_dict[line['CLIENT_CITY'].lower()] += 1

# in principle they can be unmatched
for city in velocities_up_dict.keys():
    print "At city '%s' the mean dl speed is %g, while the mean upload speed is %g" % (
        city.title(),
        velocities_up_dict[city]/data_len_by_city_dict[city],
        velocities_down_dict[city]/data_len_by_city_dict[city],
    )
```

## EX 7 

Find the more frequent **ISP**

```python
isp_c = Counter([line['ISP'].lower() for line in data])
print "The most common ISP is '%s' (with %s entries out of %s)" % (
    isp_c.most_common(1)[0][0],
    isp_c.most_common(1)[0][1],
    sum(isp_c.values())
)
```

## EX 8

Find the more frequent **BROWSER**

```python
browser_c = Counter([line['CLIENT_BROWSER'].lower().split()[0] for line in data])
print "The most common browser is '%s' (with %s entries)" % (browser_c.most_common(1)[0][0], browser_c.most_common(1)[0][1])
```

## EX 9

For the *more frequent* ISP:

* Find entry with MAX / MIN Value of DOWNLOAD_KBPS
* Find more frequent City
* Find more frequent Browser

```python
import uuid
## add uuid
for line in data:
    line['uuid'] = str(uuid.uuid4())

import pprint
    
most_frequent_isp_name = isp_c.most_common(1)[0][0]

clients_with_most_frequent_isp = filter(lambda x: x['ISP'].lower() == most_frequent_isp_name, data)
max_dwnld_most_frequent_isp = max(int(entry['DOWNLOAD_KBPS']) for entry in clients_with_most_frequent_isp)
min_dwnld_most_frequent_isp = min(int(entry['DOWNLOAD_KBPS']) for entry in clients_with_most_frequent_isp)

print "For the clients with the most common ISP, the most frequent city is '%s', and the most frequent browser is '%s'" % (
    Counter(c['CLIENT_CITY'].lower() for c in clients_with_most_frequent_isp).most_common(1)[0][0].title(),
    Counter(c['CLIENT_BROWSER'].split()[0] for c in clients_with_most_frequent_isp).most_common(1)[0][0].title()
)
print "-".center(80, '-')

for line in data:
    if int(line['DOWNLOAD_KBPS']) == max_dwnld_most_frequent_isp:
        print "The entry with the MAX value of DOWNLOAD_KBPS :"
        pprint.pprint(line)
        print "-".center(80, '-')
    if int(line['DOWNLOAD_KBPS']) == min_dwnld_most_frequent_isp:
        print "The entry with the MIN value of DOWNLOAD_KBPS :"
        pprint.pprint(line)
```

## EX 10

For each Operatin system, find the more frequent Browser

```python
for operating_system in Counter(line['CLIENT_OPERATING_SYSTEM'] for line in data):
    clients_with_this_os = filter(lambda x: x['CLIENT_OPERATING_SYSTEM'] == operating_system, data)
    c = Counter(x['CLIENT_BROWSER'].split()[0].lower() for x in clients_with_this_os)
    most_common_browser_name, most_common_browser_count = c.most_common(1)[0]
    print "For OS '%s' the most common browser is '%s' with %s occurancies out of %s, which is %g%%" % (
        operating_system.title(),
        most_common_browser_name.title(),
        most_common_browser_count,
        len(clients_with_this_os),
        float(most_common_browser_count) / len(clients_with_this_os) * 100
    )
```
