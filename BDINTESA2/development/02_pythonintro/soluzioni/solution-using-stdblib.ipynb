## Esercizio 1

Create a function to print each element of a given list:

    POSITION X --> VALUE
    POSITION Y --> VALUE

Where X and Y will be, respectively, the position of the element in the list.

```python
def print_list(lista):
    for index, value in enumerate(lista):
        print 'POSITION {} --> VALUE {}'.format(index, value)
print_list(['hello', 1, {'Ciao':True}, True])
```

## Esercizio 2
Create a function to print each element of a dictionary:

    KEY X --> VALUE
    KEY Y --> VALUE

Where X and Y will be, respectively, the keys of the dictionary.

```python
def print_dict(dizionario):
    for k,v in dizionario.items():
        print 'KEY {} --> VALUE {}'.format(k, v)
print_dict({'A':0, 'B':1, 'C':2})
```

```python
# EX 5
turin_downloads = [float(entry['DOWNLOAD_KBPS']) for entry in data if entry['CLIENT_CITY'] == 'Turin']
turin_uploads = [float(entry['UPLOAD_KBPS']) for entry in data if entry['CLIENT_CITY'] == 'Turin']

total = 0
for x in turin_downloads:
    total += x
print "DOWNLOAD MEAN FOR TURIN: ", total / len(turin_downloads)

total = 0
for x in turin_uploads:
    total += x
print "UPLOAD MEAN FOR TURIN: ", total / len(turin_uploads)
```

```python
# EX 6
cities = {}
for entry in data:
    if entry['CLIENT_CITY'] not in cities:
        cities[entry['CLIENT_CITY']] = {'DOWNLOAD': [], 'UPLOAD': []}
    cities[entry['CLIENT_CITY']]['DOWNLOAD'].append(float(entry['DOWNLOAD_KBPS']))
    cities[entry['CLIENT_CITY']]['UPLOAD'].append(float(entry['UPLOAD_KBPS']))

#print cities
for city, obj in cities.items():
    print "CITY: {}\n\tDOWNLOAD: {}\n\tUPLOAD: {}\n\n".format(city, 
                                                              sum(obj['DOWNLOAD'])/len(obj['DOWNLOAD']),
                                                              sum(obj['UPLOAD'])/len(obj['UPLOAD']))
```

```python
# EX 7
isp_list = {}
for entry in data:
    if entry['ISP'] not in isp_list:
        isp_list[entry['ISP']] = 0
    isp_list[entry['ISP']] += 1

print isp_list
isp_max = {'ISP': None, 'VALUE': -1}
for isp, value in isp_list.items():
    if value > isp_max['VALUE']:
        isp_max['ISP'] = isp
        isp_max['VALUE'] = value

print '\n'
print isp_max
```

```python
# EX 8
browser_list = {}
for entry in data:
    if entry['CLIENT_BROWSER'] not in browser_list:
        browser_list[entry['CLIENT_BROWSER']] = 0
    browser_list[entry['CLIENT_BROWSER']] += 1

print browser_list
browser_max = {'CLIENT_BROWSER': None, 'VALUE': -1}
for browser, value in browser_list.items():
    if value > browser_max['VALUE']:
        browser_max['CLIENT_BROWSER'] = browser
        browser_max['VALUE'] = value

print '\n'
print browser_max
```
