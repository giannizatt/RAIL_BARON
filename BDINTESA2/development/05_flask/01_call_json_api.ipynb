# How to perform HTTP Requests

## Urllib2

This module provides a high-level interface for fetching data across the World Wide Web. In particular, the **urlopen()** function is similar to the built-in function open(), but accepts Universal Resource Locators (URLs) instead of filenames.



```python
import urllib2
```

```python
url = "https://www.intesasanpaolo.com"
result = urllib2.urlopen(url)
print result
```

```python
result.read()
```

```python
# Call Json API
url = "https://www.intesasanpaolo.com/content/dam/vetrina/mock/banks-list.json"
result = urllib2.urlopen(url)
result.read() # string!!
```

```python
# Call Json API
url = "https://www.intesasanpaolo.com/content/dam/vetrina/mock/banks-list.json"
result = urllib2.urlopen(url)

import json
data = json.load(result)
data
```
